```python
# help(OneHotEncoder)
```

```python
encoder = OneHotEncoder(drop = 'first', handle_unknown='ignore', sparse_output = False)
```

```python
def encoders(df):
    new = pd.DataFrame()
    for col in df.select_dtypes(include='object').columns:
        new = pd.concat(
            [new, pd.DataFrame(encoder.fit_transform(df[[col]]),
                               columns=encoder.get_feature_names_out([col]))],
            axis=1
        )

    return new
```

```python
df_encoded = encoders(df)
```
